```python
import os

def option():
    print("Enter an option: ") 
    option = (input(" \n option 1 : Register \n option 2 : Login \n option 3 : Forgot password \n "))
    if option == '1':
       return register()
    elif option == '2':
       return login()
    elif option == '3':
       return forget_password()
    else:
       return 'enter a valid option'

def database():                                     
    with open("/content/database.txt" ,"r") as fp:   # reading the txt file that contains data
         for line in fp.readlines():
             email_id, password = line.split(" ")   # splitting into email_id & password based on whitespace
             yield email_id, password[:-1]

def login():
    email_id, password = get_inputs()      # getting inputs
    if validate(email_id, password):   # validating both email_id & password in the database
       return "Login success!!!"

def validate(email_id, password):             
    return any(i == (email_id,password) for i in database())   # returns 'True' if any such id & pass exists in the database

def get_inputs():
    email_id = input("email id:")
    password = input("Password:")
    return email_id, password

def forget_password():
    email_id, password = get_inputs()
    if user_exists(email_id):         # returns 'True' if such email_id exists in database
       return user_exists(email_id)            # corresponding password is retrived

def user_exists(email_id):                
    for usr_name, passwod in database():
        if usr_name == email_id:          # if user exists in database, password is retrieved
           return "The retrived password is: "+passwod    
    else:
        return "email_id doesnot exist, please register"   

def register():
    email_id, password = get_email_password()
    if email_id == 'invalid email id':
      return 'enter a valid email id'
    if password == 'invalid password':
      return 'enter a valid password'
    elif user_exists(email_id):          # email_id already exists
      return "email_id Unavailable. Please Try Again"
    else:
       db = open("/content/database.txt","a")          # appending the database
       db.write(email_id + " " + password + os.linesep)
       print("Registration success !!!")
       db.close()

def get_email_password():        
    email_id = email_check()
    password = password_check()
    return email_id, password
       
def email_check():
    email = input("email_id : ")
    X = list(email)
    if '@' in email and '.' in email:
        diff = X.index('.')-X.index("@")   # checking index position of '@' and '.'
        d,e = 0,0
        if diff > 1:         # '@' and '.' should not be nearby
            d+=1
        if X[0].isalpha():       # start with alphabets
            e+=1
        if d==1 & e==1:
            return email
        else:
            print('email id should start with alphabets')
            return 'invalid email id'
    else:
        print('email id should contain @ and .')
        return 'invalid email id'
        
def password_check():
    print('password should have atleast one digit, one uppercase letter, one lowercase letter, ')
    print('one special characters and length should be between 5 & 16')
    passwd = input("Password: ")
    Special_char =['$', '@', '#', '%','~','`','!','*','^','&','/','-']
    val = True  
    if len(passwd) < 5 and len(passwd) > 16:
        val = False      
    if not any(item.isdigit() and  item.isupper() and item.islower() for item in passwd):
        val = False         
    if not any(item in Special_char for item in passwd):
        val = False
    if val:               # if True
        return passwd
    else:
        return 'invalid password'

option()
```
